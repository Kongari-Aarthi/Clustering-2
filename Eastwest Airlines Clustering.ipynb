```python
df.isnull().sum()
```

```python
X = df.iloc[:,1:]
X.shape
df.corr()
```

```python
from sklearn.preprocessing import StandardScaler  # Minmaxscaler
SS = StandardScaler()
X_scale = SS.fit_transform(X)
```

```python
# construction of Dendogram
import scipy.cluster.hierarchy as shc
```

```python

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 7))  
plt.title("Airlines")  
dend = shc.dendrogram(shc.linkage(X, method='complete')) 
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='complete')
Y = cluster.fit_predict(X)
```

```python
Y_new = pd.DataFrame(Y)
Y_new[0].value_counts()
pd.concat([df,Y_new],axis=1)
```

```python
# K-Means Clustering
```

```python
from sklearn.cluster import KMeans
```

```python
kint_list = []
for i in range(2,15):
    KM = KMeans(n_clusters=i, n_init=20)
    KM.fit(X_scale)
    kint_list.append(KM.inertia_)
print(kint_list)
```

```python
import matplotlib.pyplot as plt
plt.scatter(range(2,15),kint_list)
plt.plot(range(2,15),kint_list)
plt.show()
```

```python
KM = KMeans(n_clusters = 6 , n_init = 20)
KM.fit(X_scale)
Y = KM.predict(X_scale)
```

```python
y = pd.DataFrame(Y)
y.value_counts()
```

```python
# DBSCAN Clustering
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps = 2, min_samples= 3)
dbscan.fit(X_scale)
```

```python
y = dbscan.labels_
Y = pd.DataFrame(y)
Y.value_counts()
```
